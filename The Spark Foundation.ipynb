###### step-4 This sorting step is performed to more accuratly visulise the data more the study hours more the percentage.Here in this case data is already in sorted manner.This step is commented during training of the model for accuracy.

```python
dta.sort_values(["Hours"],axis=0,ascending=[True],inplace=True)
dta.head(8)
```

```python
dta.plot.bar(x='Hours',y='Scores')
```

###### step-5 Preperation of data and dividing data for test and training the model.

```python
x=dta.iloc[:,:-1].values
y=dta.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

###### step-6 Training the algorithm

```python
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
regr.fit(x_train,y_train)
print("training is complete")
```

```python
line=regr.coef_*x+regr.intercept_
plt.scatter(x,y,color='green')
plt.plot(x,line,color='red')
plt.title('Hours vs Scores')
plt.xlabel('study hours')
plt.ylabel('Percentage score')
plt.show()
```

###### step-7 Testing the algorithm

```python
print(x_test)
print("score pridiction")
y_p=regr.predict(x_test)
print(y_p)
```

###### step-8 checking accuracy of the model

```python
df=pd.DataFrame({'Actual':y_test,'Predict':y_p})
df
```

```python
df.plot(kind='line')
plt.show()
```

###### step-9 Checking with custom input

```python
hour=[[8.6]]
pr=regr.predict(hour)
print(pr)
```

###### step-10 Evaluation of the model

```python
from sklearn import metrics
print('mean absolute error:',metrics.mean_absolute_error(y_test,y_p))
print('mean squared error:',metrics.mean_squared_error(y_test,y_p))
```
